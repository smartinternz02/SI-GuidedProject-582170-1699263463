```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
# 1. Histogram - Distribution of 'total' (Number of drivers involved in fatal collisions per billion miles)
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
sns.histplot(car_crashes['total'], bins=20, kde=True, color='coral')
plt.title("Distribution of Number of Drivers Involved in Fatal Collisions")
```

```python
# 2. Histogram - Distribution of 'speeding' (Percentage of drivers involved in fatal collisions who were speeding)
plt.subplot(2, 3, 2)
sns.histplot(car_crashes['speeding'], bins=20, kde=True, color='lightblue')
plt.title("Distribution of Percentage of Speeding Drivers in Fatal Collisions")
```

```python
# 3. Histogram - Distribution of 'alcohol' (Percentage of drivers involved in fatal collisions who were alcohol-impaired)
plt.subplot(2, 3, 3)
sns.histplot(car_crashes['alcohol'], bins=20, kde=True, color='lightgreen')
plt.title("Distribution of Percentage of Alcohol-Impaired Drivers in Fatal Collisions")
```

```python
# 4. Box Plot - Car Insurance Premiums
plt.subplot(2, 3, 4)
sns.boxplot(y='ins_premium', data=car_crashes, palette='muted')
plt.title("Box Plot of Car Insurance Premiums")
```

```python
# 5. Box Plot - Losses Incurred by Insurance Companies
plt.subplot(2, 3, 5)
sns.boxplot(y='ins_losses', data=car_crashes, palette='pastel')
plt.title("Box Plot of Losses Incurred by Insurance Companies")
```

```python
# 6. Bar Plot - Car Crashes by State (Abbreviation)
state_crashes = car_crashes.groupby('abbrev')['total'].mean().reset_index()
plt.subplot(2, 3, 6)
sns.barplot(x='abbrev', y='total', data=state_crashes, palette='Set2')
plt.title("Average Car Crashes by State")
plt.xticks(rotation=45)
```

```python
# 7. Scatter Plot - 'total' vs. 'speeding'
plt.subplot(3, 3, 7)
sns.scatterplot(x='total', y='speeding', data=car_crashes, color='purple')
plt.title("Number of Drivers vs. Percentage of Speeding Drivers")
```

```python
# 8. Scatter Plot - 'total' vs. 'alcohol'
plt.subplot(3, 3, 8)
sns.scatterplot(x='total', y='alcohol', data=car_crashes, color='orange')
plt.title("Number of Drivers vs. Percentage of Alcohol-Impaired Drivers")
```

```python
# 9. Pair Plot - Pairwise relationships between numeric columns
plt.subplot(3, 3, 9)
sns.pairplot(car_crashes, vars=['total', 'speeding', 'alcohol', 'ins_premium', 'ins_losses'], diag_kind='kde', hue='abbrev')
plt.suptitle("Pairwise Relationships and Distributions")
plt.subplots_adjust(top=0.9)
```
